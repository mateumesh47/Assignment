```python
#5
n=100
x=np.random.randn(n)
y=0.8*x+np.random.randn(n)*0.5
z=0.5*x+0.5*y+np.random.randn(n)*0.5

df=pd.DataFrame({'X':x,'Y':y,'Z':z})
mat=df.corr()
sns.heatmap(mat)
```

```python
#2
np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df= pd.DataFrame(data)
px.violin(df, y='Score', x='Grade')
```

```python
#3
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

hd=df.pivot_table(index='Day',columns='Month',values='Sales')
px.imshow(hd)
```

```python
#4
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

import plotly.graph_objects as go
go.Figure(data=[go.Surface(z=z,x=x,y=y)])
```

```python
#bokeh
#1
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
x=np.linspace(0, 10, 100)
y=np.sin(x)
output_notebook()

p=figure(title="Sine Wave",x_axis_label='x',y_axis_label='sin(x)')
p.line(x,y,line_width=2)
show(p)
```

```python
#2
import numpy as np
import bokeh.plotting as bp

x=np.random.rand(100)
y=np.random.rand(100)
sizes=np.random.randint(10,200,size=100)
colors=np.random.rand(100,3)
output_notebook()

fig=bp.figure()
fig.scatter(x,y,size=sizes,color=colors)
bp.show(fig)
```
