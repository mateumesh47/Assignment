```python
#14
date=pd.date_range(start='2024-01-01',periods=30,freq='D')
temperature=np.random.uniform(low=15,high=30,size=30)
humidity=np.random.uniform(low=40,high=90,size=30)

df=pd.DataFrame({
    'Date':date,
    'Temperature':temperature,
    'Humidity':humidity
})

fig, ax1=plt.subplots()

ax1.plot(df['Date'],df['Temperature'],'r-',label='Temperature')
ax1.set_ylabel('Temperature (Â°C)',color='red')
ax1.set_xlabel('Date')


ax2 = ax1.twinx()
ax2.plot(df['Date'],df['Humidity'],'b-',label='Humidity')
ax2.set_ylabel('Humidity (%)',color='blue')

plt.title('Temperature and Humidity Over Time')
```

```python
#15 16
from scipy.stats import norm

data=np.random.normal(loc=0,scale=1,size=1000)
plt.hist(data,bins=30,density=True,alpha=0.6,color='b',label='Histogram')

xmin,xmax=plt.xlim()
x=np.linspace(xmin,xmax,100)
p=norm.pdf(x,np.mean(data),np.std(data))
plt.plot(x,p,'k',linewidth=2,label='PDF')

plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
```

```python
#17
import seaborn as sns

x=np.random.randn(100)
y=np.random.randn(100)

df=pd.DataFrame({'X':x,'Y':y})
df['Quadrant'] = np.where((df['X'] >= 0) & (df['Y'] >= 0),'Quadrant 1',
                   np.where((df['X'] < 0) & (df['Y'] >= 0),'Quadrant 2',
                   np.where((df['X'] < 0) & (df['Y'] < 0),'Quadrant 3','Quadrant 4')))

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='X',y='Y',hue='Quadrant',palette='viridis',s=100,alpha=0.7)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrant')

plt.show()
```

```python
#18
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()

x=np.linspace(0,2*np.pi,100)
y=np.sin(x)

p=figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)',width=800,height=400)
p.line(x,y,line_width=2,color='blue',legend_label='Sine Wave')

p.grid.grid_line_color='gray'
p.grid.grid_line_alpha=0.5

show(p)
```

```python
#19
from bokeh.models import ColumnDataSource, HoverTool
output_notebook()

categories=['Category A','Category B','Category C','Category D','Category E']
values=np.random.randint(10,100,size=len(categories))

data=pd.DataFrame({'Category':categories,'Value': values})
p=figure(x_range=data['Category'],title="Random Categorical Bar Chart",x_axis_label='Category',y_axis_label='Value',width=800,height=400)
p.vbar(x='Category',top='Value',source=data,width=0.5,color='blue',legend_label='Values')

hover = HoverTool()
hover.tooltips = [("Category","@Category"),("Value","@Value")]
p.add_tools(hover)

show(p)
```
